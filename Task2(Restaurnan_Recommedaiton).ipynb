```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/drive/MyDrive/almabetter 1/datasets/Dataset .csv')
df.head()
```

```python
df.info()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Fill missing values in 'Cuisines' with 'Unknown'
df['Cuisines'].fillna('Unknown', inplace=True)

# Verify that there are no more missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Select the columns to be encoded
categorical_columns = ['City', 'Cuisines', 'Currency', 'Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu', 'Rating color']

# Apply one-hot encoding to these columns
data_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

# Display the first few rows of the encoded dataset
data_encoded.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Extract features for similarity calculation
features = ['Aggregate rating'] + [col for col in data_encoded.columns if 'Cuisines_' in col]
data_features = data_encoded[features]

# Calculate cosine similarity
cosine_sim = cosine_similarity(data_features, data_features)

# Function to get recommendations based on restaurant index
def get_recommendations(index, cosine_sim=cosine_sim):
    sim_scores = list(enumerate(cosine_sim[index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Get top 10 recommendations
    restaurant_indices = [i[0] for i in sim_scores]
    return df.iloc[restaurant_indices]

# Example: Get recommendations for the first restaurant
recommendations = get_recommendations(0)
recommendations[['Restaurant Name', 'Cuisines', 'Aggregate rating']]
```

```python
# Define sample user preferences
user_preferences = {'Cuisines': 'Italian', 'Aggregate rating': 4.0}

# Function to recommend restaurants based on user preferences
def recommend_restaurants(user_preferences, data=df):
    filtered_data = df[(df['Cuisines'].str.contains(user_preferences['Cuisines'])) & (df['Aggregate rating'] >= user_preferences['Aggregate rating'])]
    return filtered_data[['Restaurant Name', 'Cuisines', 'Aggregate rating']].head(10)

# Get recommendations
user_recommendations = recommend_restaurants(user_preferences)
user_recommendations
```
