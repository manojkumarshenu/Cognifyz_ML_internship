```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/almabetter 1/datasets/Dataset .csv')
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Fill missing values in 'Cuisines' with 'Unknown'
df['Cuisines'].fillna('Unknown', inplace=True)

# Verify that there are no more missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Select the columns to be encoded
categorical_columns = ['City', 'Cuisines', 'Currency', 'Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu', 'Rating color']

# Apply one-hot encoding to these columns
data_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

# Display the first few rows of the encoded dataset
data_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Define the features (X) and target (y)
X = data_encoded.drop(columns=['Aggregate rating'])
y = data_encoded['Aggregate rating']

categorical_cols = X.select_dtypes(include=['uint8']).columns  # Assuming one-hot encoded columns are of type 'uint8'
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns

# Create a ColumnTransformer to handle categorical and numerical features
transformer = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
        ('num', 'passthrough', numerical_cols)
    ]
)

# Apply the transformations
X_transformed = transformer.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, random_state=42)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Initialize the model
lr = LinearRegression()

# Train the model
lr.fit(X_train, y_train)

# Make predictions
y_pred = lr.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
# Get the feature importances
feature_importances = pd.Series(lr.coef_, index=transformer.get_feature_names_out()) # Get feature names from the transformer
print(feature_importances.sort_values(ascending=False).head(10))
```
